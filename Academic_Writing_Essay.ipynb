```python
class Rectangle:
    def __init__(self, width=1, height=1):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return self.width * 2 + self.height * 2

    def is_square(self):
        return self.width == self.height
```

```python
r = Rectangle(4, 6)

print("Area of rectangle: ", r.area())
print("Perimeter of rectangle: ", r.perimeter())
print("Is rectangle a square? ", r.is_square())
```

## Linguistic analysis

Let's now try to conduct a short liguistic analysis based on data scraped from the Internet. I'll be using __Python__ and __NLTK__.

```python
import requests
```

```python
page = requests.get("https://www.ibm.com/think/topics/natural-language-processing")
print(page.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
print(soup)
```

```python
paras = soup.find_all("p")
print(paras)
```

```python
for el in paras:
  print(el)
```

```python
for el in paras:
  print(el.getText())
```

```python
only_text_list = []
for el in paras:
  text = el.getText().strip()
  if len(text) > 0:
    only_text_list.append(text)

print(only_text_list)
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
tuples = []

for sentence in only_text_list:
  tokenized = nltk.word_tokenize(sentence)
  print(tokenized)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
tuples = []

for sentence in only_text_list:
  tokenized = nltk.word_tokenize(sentence)
  tagged = nltk.pos_tag(tokenized)
  for item in tagged:
    tuples.append(item)

print(tuples)
```

```python
counter_dict = {}
for item in tuples:
    print(item)
    if item[1] not in counter_dict:
        counter_dict[item[1]] = 1
    else:
        counter_dict[item[1]] += 1

print(counter_dict)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

keys = []
values = []
for el in counter_dict:
    keys.append(el)
    values.append(counter_dict[el])

print(keys)
print(values)
```

```python
plt.bar(keys[0:10], values[0:10], color = "xkcd:sky blue")
```

```python
plt.pie(values[0:10], labels=keys[0:10])
```

```python
word_counter = {}
for item in tuples:
    word = item[0]
    if word not in word_counter:
        word_counter[word] = 1
    else:
        word_counter[word] += 1

print(word_counter)
```

```python
from collections import defaultdict, Counter

target_tags = {'JJ', 'NNP', 'NN', 'VBZ', 'DT', 'IN', 'CC'}

tagged_word_counts = defaultdict(Counter)

for word, tag in tuples:
    if tag in target_tags:
        normalized_word = word.lower()
        tagged_word_counts[tag][normalized_word] += 1

for tag in target_tags:
    print(f"\nTop 5 words for '{tag}' tag:")
    for word, count in tagged_word_counts[tag].most_common(5):
        print(f"{word}: {count}")
```
